```python
import numpy as np
import pandas as pd
np.set_printoptions(precision=4)


df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
df
```

```python
#df['FlightNumber'][1]= df['FlightNumber'].fillna(value=df.iloc[1][1]+10)
df.fillna(method='ffill', inplace=True)
df.at[1,'FlightNumber']= df["FlightNumber"][0] +10
df.at[3,'FlightNumber']= df["FlightNumber"][2] +10
df['FlightNumber'] = df['FlightNumber'].astype(int)
df
```

```python
new = df["From_To"].str.split("_", n = 1, expand = True) 

df["From"]= new[0] 
df["To"]= new[1] 
df['From'] = df['From'].str.capitalize()
df['To'] = df['To'].str.capitalize()
df.drop(columns =['From_To'], inplace = True)
df
```

```python
df1= pd.DataFrame(df["RecentDelays"].to_list(), columns=['Delay1', 'Delay2','Delay3'])
df1
df["Delay1"]=df1["Delay1"]
df["Delay2"]=df1["Delay2"]
df["Delay3"]=df1["Delay3"]
df.drop(columns =['RecentDelays'], inplace = True)
df
```
